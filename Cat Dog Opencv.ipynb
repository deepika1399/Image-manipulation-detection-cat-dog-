```python
os.chdir(r"C:\Users\ravin\Desktop\cat_train")
all_images=os.listdir()
cat=[]
for i in all_images:
    img1=cv2.imread(r"C:\Users\ravin\Desktop\cat_train\\"+i)
    grey=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY) #changing colorspace
    resized=cv2.resize(grey,(150,100))  #here am reducing the pixels
    f_cat=resized.flatten().tolist()
    cat.append(f_cat)
```

```python
cat_data=pd.DataFrame(cat)
cat_data["label"]="cat"
```

```python
os.chdir(r"C:\Users\ravin\Desktop\dog_train")
all_images=os.listdir()
dog=[]
for dog_images in all_images:
    img1=cv2.imread(r"C:\Users\ravin\Desktop\dog_train\\"+dog_images)
    grey=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
    resize_=cv2.resize(grey,(150,150))
    f_dog=resize_.flatten().tolist()
    dog.append(f_dog)
```

```python
dog_data=pd.DataFrame(dog)
dog_data["label"]="dog"
```
